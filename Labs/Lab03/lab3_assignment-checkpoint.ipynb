### Discussion 1.1

Just from the simple descibre statistics above, it is clear that **_departure_**, **_arrival_**, and **_late aircraft_** delays on average are longer in July. The reason why this is the case is related to the physics behind how airplanes take off. During warmer temperatures, the air is thinner, which reduces the lift required for the airplane to take off. Due to this issue, some pilots may elect to delay flights, dump fuel, or have less lugggage on board to reduce the weight of the aircraft [Source - 1].

Another possible reason is because Summer is travel season, so the large number of people may be flying simultaneously, causing logistics at airports to be more difficult.

The fact that more planes are delayed compared to arriving early is not a surprise, as often times unforseen things occur causing the flights to be delayed. It also makes sense from a logistics stand point. From the airliners' point of view, as if the mean of delays are below 0, meaning they arrive early, then they could be getting more out of these planes by making them fly more frequently. So airlines would rather the planes be delayed rather having them arrive early and being idle.

### Discussion 1.2

```python
# 2. Delay Investigation with Days of the Week
delay_data1_with_date = data1[['DayOfWeek', 'Reporting_Airline', 'DepDelay', 'ArrDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay', "TotalAddGTime"]]
delay_data7_with_date = data7[['DayOfWeek', 'Reporting_Airline', 'DepDelay', 'ArrDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay', "TotalAddGTime"]]

# Clear data of any NaNs
dd1_cleaned_with_date = delay_data1_with_date.dropna(axis=1, how='all')
dd7_cleaned_with_date = delay_data7_with_date.dropna(axis=1, how='all')

dd1_cleaned_with_date = dd1_cleaned_with_date.dropna(axis=0, how='any')
dd7_cleaned_with_date = dd7_cleaned_with_date.dropna(axis=0, how='any')

print(f"Number of rows in cleaned data1 with date: {len(dd1_cleaned_with_date)}, Number of columns in cleaned data1 with date: {len(dd1_cleaned_with_date.columns)}")
print(f"Number of rows in cleaned data7 with date: {len(dd7_cleaned_with_date)}, Number of columns in cleaned data7 with date: {len(dd7_cleaned_with_date.columns)}")
```

```python
# Construct historical mean delay time for each day of the week
day_of_week_data1 = dd1_cleaned_with_date.groupby('DayOfWeek').mean()
day_of_week_data7 = dd7_cleaned_with_date.groupby('DayOfWeek').mean()

# Map 1-7 to Monday-Sunday
day_of_week_data1.index = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
day_of_week_data7.index = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Plot the data
fig, ax = plt.subplots(2, 1, figsize=(10, 10))
day_of_week_data1.plot(kind='bar', ax=ax[0])
day_of_week_data7.plot(kind='bar', ax=ax[1])
ax[0].set_title("Average Delay Time by Day of the Week (January)")
ax[1].set_title("Average Delay Time by Day of the Week (July)")
plt.tight_layout()
plt.show()
```

# Discussion 2.1
By looking at the average time of delay by days of the week, there are no outstanding days that has a delay spike. However, it can be seen that _Monday_ and _Saturday_ generally have longer delay times compared to the rest of the days. This makes sense as usually Monday is the first business day of the week, and a lot of peopel travel on that day instead of Sunday so they can spend time with their family. On Saturday, that is usually when people are able to travel since it is the first day of the weekend, so it is also not a surprise that this day has more delays than others as the quantity of flights increases.

However, these graphs do not tell the full story, as the day-to-day delay difference is mostly within 15 minutes of each other. Additional data from other months is needed to have a more conclusive result, especially those not during holiday/break seasons, which Janurary and July are.

```python
# Construct historical total delay time for each day of the week for each type of delay
day_of_week_data1_total = dd1_cleaned_with_date.groupby('DayOfWeek').sum()
day_of_week_data7_total = dd7_cleaned_with_date.groupby('DayOfWeek').sum()

# Map 1-7 to Monday-Sunday
day_of_week_data1_total.index = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
day_of_week_data7_total.index = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Plot the data
fig, ax = plt.subplots(2, 1, figsize=(10, 10))
day_of_week_data1_total.plot(kind='bar', ax=ax[0])
day_of_week_data7_total.plot(kind='bar', ax=ax[1])
ax[0].set_title("Total Delay Time by Day of the Week (January)")
ax[1].set_title("Total Delay Time by Day of the Week (July)")
plt.tight_layout()
plt.show()
```
